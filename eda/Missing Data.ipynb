```python
import numpy as np
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./PenguinsClean.csv')
df.head()
```

```python
df.shape
```

```python
# filter out chinstraps
df = df[(df['species'] == 'Gentoo') | (df['species'] == 'Adelie')]
df.shape
```

```python
from pycaret.classification import *
setup(df, target='sex');
```

```python
compare_models()
```

```python
ridge_model = create_model('ridge', fold=10)
```

```python
lda_model = create_model('lda', fold=10)
```

```python
catboost_model = create_model('catboost', fold=10)
```

```python
evaluate_model(lda_model)
```

```python
evaluate_model(catboost_model)
```

```python
unseen = pd.read_csv('./penguins_holdouts.csv')
unseen = unseen.drop(['sex'], axis=1)
unseen
```

```python
unseen_pred1 = predict_model(ridge_model, data=unseen)
unseen_pred1
```

```python
unseen_pred2 = predict_model(lda_model, data=unseen)
unseen_pred2.sort_values('prediction_score', ascending=False)
```

```python
unseen_pred3 = predict_model(catboost_model, data=unseen)
unseen_pred3.sort_values('prediction_score', ascending=False)
```
